# Customer Churn Prediction (Supervised Learning - Classification)
- Problem: Predict whether a customer is likely to stop purchasing (churn).

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, r2_score, mean_absolute_error, mean_absolute_error
```

```python
x = df.drop(columns=['Order Status','Cluster','Purchase Date'])
y = df['Order Status']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=.2,random_state=42)
```

```python
print(x_train.shape,'\n')
print(x_test.shape)
```

```python
print(y_train.shape,'\n')
print(y_test.shape)
```

```python
# Decision Tree
Dec_tree = DecisionTreeClassifier()
Dec_tree.fit(x_train,y_train)
y_pred = Dec_tree.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
# Random Forest
model_RF = RandomForestClassifier()
model_RF.fit(x_train,y_train)
y_pred= model_RF.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
Lr = LogisticRegression()
Lr.fit(x_train,y_train)
y_pred= Lr.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
# xgboost
import xgboost as xgb
model_xgb= xgb.XGBClassifier()
model_xgb.fit(x_train,y_train)
y_pred= model_xgb.predict(x_test)
accuracy_score(y_test,y_pred)
```

# Product Recommendation (Association Rule Mining)
- Problem: Recommend products to customers based on their purchase history.

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv('Electronic_sales_Sep2023-Sep2024.csv')
basket = df.groupby(['Customer ID','Product Type']).size().unstack(fill_value=0)
```

```python
basket
```

```python
basket.max()
```

```python
basket= basket.applymap(lambda x:1 if x >0 else 0)
```

```python
basket.max()
```

```python
frequest_items = apriori(basket,min_support=0.05,use_colnames=True)
frequest_items
```

```python
rules = association_rules(frequest_items, metric='lift', min_threshold=.5)
print(rules)
```

# Demand Forecasting (Time Series Prediction)
- Problem: Forecast future sales of products based on historical data.

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
df['Purchase Date']= pd.to_datetime(df['Purchase Date'])
```

```python
df_ts = df.groupby('Purchase Date')['Quantity'].sum()
# df_ts
```

```python
model_arima= ARIMA(df_ts,order=(5,1,0))
mod_fit=model_arima.fit()
forcast = mod_fit.forecast(steps=30)
print(forcast)
```

```python
plt.figure(figsize=(10,4))
sns.lineplot(forcast)
plt.xticks(rotation=90)plt.show()
```

# Payment Method Prediction (Classification)
- Problem: Predict the Payment Method a customer will use based on demographics and purchase behavior.

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
lr = LabelEncoder()
df['Gender']= lr.fit_transform(df['Gender'])
df['Loyalty Member']= lr.fit_transform(df['Loyalty Member'])
df['Order Status']= lr.fit_transform(df['Order Status'])
```

```python
x = df[['Age', 'Gender', 'Loyalty Member', 'Total Price', 'Quantity']]  # Features
y = df['Payment Method']  # Target
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Random Forest
model = RandomForestClassifier()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

print('Accuracy:', accuracy_score(y_test, y_pred))
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
Lr = LogisticRegression()
Lr.fit(x_train,y_train)
y_pred= Lr.predict(x_test)
print(accuracy_score(y_test,y_pred))
```

```python
# Decision Tree
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred = dt.predict(x_test)
print(accuracy_score(y_test,y_pred))
```
